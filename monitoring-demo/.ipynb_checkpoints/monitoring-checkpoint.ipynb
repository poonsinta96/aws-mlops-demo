### Setting up of Sagemaker model monitor

```python
sagebucket = 'sagemaker-project-p-kjjql6k63ho9'
endpoint_name = 'pipeline-demo-1-staging'
predictor = RealTimePredictor(endpoint=endpoint_name, content_type = 'text/csv')


model_prefix = prefix + "/" + endpoint_name
baseline_prefix = model_prefix + '/baselining'
baseline_results_prefix = baseline_prefix + '/results'

baseline_results_uri = 's3://{}/{}'.format(sagebucket, baseline_results_prefix)
train_data_header_location = "monitoring_demo/processedData/"
print('Baseline results uri: {}'.format(baseline_results_uri))
print(train_data_header_location)
```

```python
from sagemaker.model_monitor import DefaultModelMonitor
from sagemaker.model_monitor.dataset_format import DatasetFormat

my_default_monitor = DefaultModelMonitor(
    role=role,
    instance_count=1,
    instance_type='ml.m5.xlarge',
    volume_size_in_gb=20,
    max_runtime_in_seconds=3600)

my_default_monitor.suggest_baseline(
    baseline_dataset=os.path.join(train_data_header_location, 'test_with_header.csv'),
    dataset_format=DatasetFormat.csv(header=True),
    output_s3_uri=baseline_results_uri,
    wait=True
)
```

```python
s3_client = boto3.Session().client('s3')
result = s3_client.list_objects(Bucket=sagebucket, Prefix=baseline_results_prefix)
report_files = [report_file.get("Key") for report_file in result.get('Contents')]
print("Found Files:")
print("\n ".join(report_files))

baseline_job = my_default_monitor.latest_baselining_job
schema_df = pd.json_normalize(baseline_job.baseline_statistics().body_dict["features"])
schema_df
```

### Testing with gradual drift (Longterm)


```python
import time 
t0 = time.time()
time.sleep(10)
duration = time.time()-t0
d_hour = duration/60/60
print(int(d_hour * 60 *60))
time.time()
```

```python
from threading import Thread
import math

runtime_client = boto3.client('runtime.sagemaker')
ep_name = 'pipeline-demo-1-staging'

def create_gradual_drift_data(t0):
    #creating fake data
    fake_data = pd.read_csv('monitoring_demo/processedData/test_with_header.csv')

    duration_sec = time.time()-t0
    d_hour = duration_sec/60/60

    #creating a shift of +20 age in twelve hours
    shift = d_hour * 20 / 12
    
    fake_data['AGE'] = fake_data['AGE']+int(shift)
    print(shift)
    fake_data.to_csv('monitoring_demo/processedData/fake_data.csv',index = False)
    
    
# (just repeating code from above for convenience/ able to run this section independently)
def invoke_endpoint(ep_name, file_name, runtime_client):
    
    with open(file_name, 'r') as f:
        
        print(next(f))
        count = 0
        for row in f:
            payload = row.rstrip('\n')
            response = runtime_client.invoke_endpoint(EndpointName=ep_name,
                                            ContentType='text/csv', 
                                            Body=payload[4:])
            sleep(1)
    
    sleep(10)
            
def invoke_endpoint_forever():
    t0 = time.time()
    while True:
        create_gradual_drift_data(t0)
        invoke_endpoint(ep_name, 'monitoring_demo/processedData/fake_data.csv', runtime_client)

thread = Thread(target = invoke_endpoint_forever)
thread.start()
```
